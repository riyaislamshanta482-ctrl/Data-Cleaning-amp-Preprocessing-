```python
#1.Run df.isnull().sum() to count missing values per column.
# for this problem we use isnull method (it drop the vale<%5)
# import pandas and numpy
import pandas as pd
import numpy as np
df=pd.read_csv("/content/dirty_dataset.csv") # for read the dataset
#copy () the data set
df2=df.copy()
df2
df2.isnull().sum()
df2.shape
```

**2. Numeric columns (age, salary, price, weight_kg): fill with median.**

```python
#Numeric columns (age, salary, price, weight_kg): fill with median.
import pandas as pd
import numpy as np
df2 = pd.read_csv("/content/dirty_dataset.csv")

# Convert specified columns to numeric, coercing errors to NaN
#salray convert nummerical value
df2["salary"]=pd.to_numeric(df2["salary"],errors="coerce")
df2["price"] = pd.to_numeric(df2["price"], errors='coerce')
df2["weight_kg"] = pd.to_numeric(df2["weight_kg"], errors='coerce')

# Fill missing values (NaNs) in these numeric columns with their respective medians
df2["age"] = df2["age"].fillna(df2["age"].median())
df2["salary"] = df2 ["salary"] .fillna(df2["salary"].median())
df2["price"]= df2 ["price"] .fillna (df2["price"].median())
df2["weight_kg"] = df2 ["weight_kg"].fillna(df2["weight_kg"].median())
display(df2.head(10))
```

**Categorical columns (department, gender, country, city): fill with mode.**
for Categorical columns we use mode method

```python
#Categorical columns (department, gender, country, city): fill with mode.
#Fill missing values (NaNs) in these numeric columns with their respective mode
df2["department"]=df2["department"].fillna(df2["department"].mode()[0])
df2["gender"] = df2["gender"] .fillna (df2["gender"] .mode()[0])
df2["country"] = df2["country"].fillna(df2["country"].mode()[0])
df2["city"] = df2["city"].fillna(df2["city"]. mode() [0])
display(df2.head())
```

4. Date column (join_date): forward fill or fill with most frequent date.

```python
df2["join_date"] = df2["join_date"].ffill() # using fillna method
df2["join_date"]
```

5. review column: fill with No Review

```python
df2["review"] = df2["review"].fillna('No Review')
df2["review"]
```
