5. Outlier Detection & Handling
Problem Detection

Methods:

Boxplot

IQR

Z-Score

Domain Rules

Example:

Age < 0 Salary > 10M

Solutions

Remove

Cap (Winsorization)

Log Transform

````python
#Experience_Years
Q1 = df2["Experience_Years"].quantile(0.25)
Q3 = df2["Experience_Years"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

# Detect
outliers = df2[
    (df2["Experience_Years"] < lower) |
    (df2["Experience_Years"] > upper)
]

# Remove
df2 = df2[
    (df2["Experience_Years"] >= lower) &
    (df2["Experience_Years"] <= upper)
]

# Cap
df2["Experience_Years"] = df2["Experience_Years"].clip(
    lower,
    upper
)

# Log Transform
df2["Experience_Years"] = np.log1p(df2["Experience_Years"])
df2["Experience_Years"]
````

**The Experience_Years data was well distributed, with no major imbalance.**

````python
## Check distribution
print(df2["Job_Risk"].value_counts())
````

````python

#oversampling
# ---------------------
from sklearn.utils import resample

majority = df2[df2.Job_Risk == "Low"]
minority = df2[df2.Job_Risk == "High"]

minority_up = resample(
    minority,
    replace=True,
    n_samples=len(majority),
    random_state=42
)

df2_balanced = pd.concat(
    [majority, minority_up]
)

df2_balanced.Job_Risk.value_counts()

# Undersampling
# ---------------------

majority_down = resample(
    majority,
    replace=False,
    n_samples=len(minority),
    random_state=42
)

df2_under = pd.concat(
    [majority_down, minority]
)

df2_under.Job_Risk.value_counts()
````
